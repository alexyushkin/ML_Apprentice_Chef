```python
# placeholders of dummy variables for domain types
data['domain_personal'] = 0
data['domain_professional'] = 0
data['domain_junk'] = 0


# looping over each email address
for index, col in data.iterrows():
    
    # splitting email domain at '@' and add '@' at the beginning
    domain = '@' + data.loc[index, 'EMAIL'].split(sep='@')[1]
    
    # filling out personal domain column
    if domain in domain_personal:
        data.loc[index, 'domain_personal'] = 1

    # filling out professional domain column
    elif domain in domain_professional:
        data.loc[index, 'domain_professional'] = 1

    # filling out junk domain column
    elif domain in domain_junk:
        data.loc[index, 'domain_junk'] = 1
    
    # print message if a domain is missing in all lists
    else:
        print('Unknown')

        
# preventing print output if it is not needed
if show_dev_output:
    # checking results
    
    print(data[['domain_personal', 'domain_professional', 
                'domain_junk']].value_counts())
```

```python
# defining a function for displaying scatterplots
def scatterplots(x, y, data, ncols = 2):
    """This function generates scatterplots and can be used for continuous 
    variables.
    Make sure matplotlib.pyplot (as plt), seaborn (as sns), and math have 
    been imported.

    PARAMETERS
    ----------
    x : list, categorical variables
    y : str, response variable
    data : DataFrame of the response and categorical variables
    ncols : int, number of columns, default value 2
    """
    
    # generating a list of color codes for each scatterplot
    my_palette = list(sns.color_palette("tab10", len(x)).as_hex())
    
    # calculating number of rows taking into account quantities of  
    # scatterplots and columns
    nrows = math.ceil(len(x) / ncols)
    
    # creating a figure and counting its size
    fig, ax = plt.subplots(figsize = (ncols * 5, nrows * 4))
    
    # preparing a visualization
    for i in range(len(x)):
        
        # developing a scatterplot
        plt.subplot(nrows, ncols, i + 1)
        sns.scatterplot(x = data[x[i]],
                        y = data[y],
                        color = my_palette[i])
        
        # adding labels
        plt.xlabel(xlabel = x[i])
        plt.ylabel(ylabel = y)
    
    # cleaning up the layout and displaying the results
    plt.tight_layout()
    plt.show()
```

```python
# defining a function for displaying join boxplots & distplots
def joingrids_box_dist(x, y, data, style='whitegrid', size=8):
    """This function generates boxplot combined with distplots and can be used 
    for categorical variables.
    Make sure matplotlib.pyplot and seaborn have been imported 
    (as plt and sns).

    PARAMETERS
    ----------
    x : str, categorical variable
    y : str, response variable
    data : DataFrame of the response and categorical variables
    style : str, style of seaborn visuals, default value 'whitegrid'
    size : int, size of figure, default value 8
    """
    
    # setting seaborn style
    sns.set_style(style)
    
    # creating joint grid plot
    g = sns.JointGrid(x=x, y=y, data=data, size=size)

    # creating boxplot combined with distplots
    g.plot(sns.boxplot, sns.distplot)

    # displaying the plot
    plt.show();
```

```python
# defining a function for displaying boxplots & distplots together in one row
def boxplot_distplot(x, y, data, nrows=1, ncols=2, figsize=(12, 4)):
    """This function generates boxplot and distplot seperately in one row and  
    can be used for count variables.
    Make sure matplotlib.pyplot and seaborn have been imported 
    (as plt and sns).

    PARAMETERS
    ----------
    x : str, categorical variable
    y : str, response variable
    data : DataFrame of the response and categorical variables
    nrows : int, number of rows, default value 1
    ncols : int, number of columns, default value 2
    figsize : tuple, size of figure, default value (12, 4)
    """
    
    # creating a plot with 1 row and 2 columns 
    fig, (ax0, ax1) = plt.subplots(nrows=nrows, ncols=ncols, 
                                   figsize=figsize)

    # creating a boxplot  
    sns.boxplot(x=x, y=y, data=data, ax=ax0)
    ax0.set(xlabel=x)

    # creating a distplot 
    sns.distplot(x=data[x], ax=ax1)
    ax1.set(xlabel=x)

    # displaying the plot
    plt.show();
```

```python
# finding the optimal number on neighbors
    
# preventing running and print output if they are not needed
if show_dev_output:

    # creating lists for training set accuracy and test set accuracy
    training_accuracy = []
    test_accuracy = []


    # building a visualization of 1 to 30 neighbors
    neighbors_settings = range(1, 31)


    for n_neighbors in neighbors_settings:

        # Building the model
        clf = KNeighborsRegressor(n_neighbors=n_neighbors)
        clf.fit(x_train, y_train)

        # Recording the training set accuracy
        training_accuracy.append(clf.score(x_train, y_train))

        # Recording the generalization accuracy
        test_accuracy.append(clf.score(x_test, y_test))


    # plotting the visualization
    fig, ax = plt.subplots(figsize=(12,8))
    plt.plot(neighbors_settings, training_accuracy, label=\
                                                     "training accuracy")
    plt.plot(neighbors_settings, test_accuracy, label="test accuracy")
    plt.ylabel("Accuracy")
    plt.xlabel("n_neighbors")
    plt.legend()
    plt.show()


    # finding the optimal number of neighbors
    opt_neighbors = test_accuracy.index(max(test_accuracy)) + 1
    print(f"""The optimal number of neighbors is {opt_neighbors}""")
```
